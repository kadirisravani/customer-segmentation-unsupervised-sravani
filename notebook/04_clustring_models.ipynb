```python

import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("../data/online_retail_customer_churn.csv")

print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns)
```

```python
df = df.drop(columns=["CustomerID"], errors="ignore")


df = df.drop(columns=["Exited"], errors="ignore")


df = pd.get_dummies(df, drop_first=True)


df = df.fillna(df.mean())
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df)

print("\nFinal Feature Matrix Shape:", X.shape)
```

```python

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X)

kmeans_sil = silhouette_score(X, kmeans_labels)
print("\nKMeans Silhouette Score:", kmeans_sil)
```

```python

hierarchical = AgglomerativeClustering(n_clusters=4)
hier_labels = hierarchical.fit_predict(X)

hier_sil = silhouette_score(X, hier_labels)
print("Hierarchical Silhouette Score:", hier_sil)
```

```python

dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

if len(set(dbscan_labels)) > 1 and -1 not in set(dbscan_labels):
    dbscan_sil = silhouette_score(X, dbscan_labels)
else:
    dbscan_sil = np.nan

print("DBSCAN Silhouette Score:", dbscan_sil)
```

```python

gmm = GaussianMixture(n_components=4, random_state=42)
gmm_labels = gmm.fit_predict(X)

gmm_sil = silhouette_score(X, gmm_labels)
print("GMM Silhouette Score:", gmm_sil)
```

```python
comparison = pd.DataFrame({
    "Model": ["KMeans", "Hierarchical", "DBSCAN", "GMM"],
    "Silhouette Score": [
        kmeans_sil,
        hier_sil,
        dbscan_sil,
        gmm_sil
    ]
})

print("\nModel Comparison:")
print(comparison.sort_values(by="Silhouette Score", ascending=False))
```
